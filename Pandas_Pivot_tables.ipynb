# Execute the code below

```
import pandas as pd
import numpy as np
link = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/weather2019.csv"
df_weather = pd.read_csv(link)
```

# You have now a DataFrame with lots of weather data

```
df_weather.head()
```

# PIVOT TABLE

## It's like in Excel, but with Pandas

There are 2 syntaxes, choose your preferred :
- As a method : YOUR_DATAFRAME.pivot_table()
- As a function : pd.crosstab()

```
# Example
df_weather.pivot_table(values = 'MAX_TEMPERATURE_C', index = 'MONTH', aggfunc = 'mean')
```

## Create a Pivot table with :
- Months on rows
- the maximum of Max Temperature on values
- Opinion on columns

```
# Hop !
df_weather.pivot_table(columns = 'OPINION' ,index = 'MONTH', values = 'MAX_TEMPERATURE_C',aggfunc = 'mean')
```

## Pivot table with :
- Days on rows
- Precipitation on values
- Months on columns


```
# Hop !

df_weather.pivot_table(columns = 'MONTH' ,index = 'DAY', values = 'PRECIP_TOTAL_DAY_MM',aggfunc = 'mean')
```

## Pivot table with :
- Months on rows
- the sum of Precipitations on values

**Then display it as a column chart.**

```
# Hop !

mask1 = df_weather.pivot_table(index = 'MONTH', values = 'PRECIP_TOTAL_DAY_MM',aggfunc = np.sum)
mask1 
```

```
import seaborn as sns
sns.set()
sns.barplot(data = mask1, x = mask1.index, y = 'PRECIP_TOTAL_DAY_MM', palette = sns.color_palette("Paired"))
```

```
df_weather.pivot_table(index="MONTH", values='PRECIP_TOTAL_DAY_MM', aggfunc='sum').plot(kind='bar')
```

## Pivot table with :
- Months on rows
- the average of Max Temperature on a first column of values
- the average of Sunhours on a second column of values

```
# Hop !

df_weather.pivot_table(index = 'MONTH', values = ['MAX_TEMPERATURE_C','SUNHOUR'],aggfunc = 'mean')
```

# Chart
Using your favorite library (MatPlotLib, Seaborn, Plotly or what you want), draw a chart from the previous pivot table with :
- Temperature in a blue column chart
- Sunhours in a red line plot

Tip : Once your pivot table created, you can help yourself with the `.reset_index()`  method.

```
# Your code here


# Your code here
import matplotlib.pyplot as plt


#ax1 = fig.add_subplot(111)
ax=df_weather.pivot_table(index='MONTH', values=['SUNHOUR'], aggfunc="mean").plot(kind='bar', colormap='Blues_r')
df_weather.pivot_table(index='MONTH', values=['MAX_TEMPERATURE_C'], aggfunc="mean").plot(kind='line',colormap="Reds_r",ax=ax)
plt.show()
```

```
plt.bar(x = df_weather["MONTH"], height= df_weather["MAX_TEMPERATURE_C"])
plt.plot(df_weather["MONTH"], df_weather["SUNHOUR"], c="red")

plt.title("Temperatures and Sunhours")
plt.legend(["Sunhour", "Max temperature"])
plt.show()
```
